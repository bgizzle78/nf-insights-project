BEA SAGDP9 WV data cleaning and exploration.

```python
# Import libraries
import pandas as pd
```

```python
# Read in data
sagdp9wv_df = pd.read_csv('../../../workspace/nf-insights-project/data_raw/BEA GDP Data/SAGDP9_WV_1997_2024.csv', low_memory=False)

print('CSV read successfully!')

sagdp9wv_df.head()
```

```python
# Data check
print(sagdp9wv_df.shape)
print(sagdp9wv_df.info())
sagdp9wv_df.columns
```

```python
# Drop unnecessary columns
years = [str(y) for y in range(2009, 2025)]

sagdp9wv_df = sagdp9wv_df[['GeoName', 'LineCode', 'IndustryClassification', 'Description', 'Unit'] + years] 

# Data check
print(sagdp9wv_df.shape)
sagdp9wv_df.info()
```

```python
# Drop completely empty rows
sagdp9wv_df = sagdp9wv_df.dropna(how='all', axis=0)

print(sagdp9wv_df.shape)
```

```python
# Melt df from wide -> long
# Columns that identify each row
id_vars = ['GeoName', 'LineCode', 'IndustryClassification', 'Description', 'Unit']

# Columns to turn into rows
value_vars = [str(y) for y in range(2009, 2025)]

sagdp9wv_long = pd.melt(
    sagdp9wv_df,
    id_vars=id_vars,
    value_vars=value_vars,
    var_name='Year',
    value_name='GDPReal'
)

print(sagdp9wv_long.shape)
print(sagdp9wv_long.info())
```

```python
# Convert LineCode from float64 to int
sagdp9wv_long['LineCode'] = sagdp9wv_long['LineCode'].astype(int)

# Convert Year from object to int
sagdp9wv_long['Year'] = sagdp9wv_long['Year'].astype(int)

# Data check
print(sagdp9wv_long.info())
sagdp9wv_long.head()
```

```python
# Strip whitespace from strings
sagdp9wv_long['Description'] = sagdp9wv_long['Description'].str.strip()
sagdp9wv_long['Unit'] = sagdp9wv_long['Unit'].str.strip()
sagdp9wv_long['GeoName'] = sagdp9wv_long['GeoName'].str.strip()

# Reset index
sagdp9wv_long = sagdp9wv_long.reset_index(drop=True)
```

```python
# Save df to csv
sagdp9wv_long.to_csv('../../../workspace/nf-insights-project/data_clean/sagdp9wv_long_clean.csv', index=False)

print('CSV saved successfully!')
```
