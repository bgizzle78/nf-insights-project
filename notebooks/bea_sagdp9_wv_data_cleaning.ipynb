BEA SAGDP9 WV data cleaning and exploration.

```python
# Import libraries
import pandas as pd

print('Libraries imported successfully!')
```

```python
# Read in data
sagdp9wv_df = pd.read_csv('../../../workspace/nf-insights-project/data_raw/BEA GDP Data/SAGDP9_WV_1997_2024.csv', low_memory=False)

print('CSV read successfully!')

sagdp9wv_df.head()
```

```python
# Data check
print('Data check:')
print('Data shape:', sagdp9wv_df.shape)
print(sagdp9wv_df.info())
sagdp9wv_df.columns
```

```python
# Drop unnecessary columns
years = [str(y) for y in range(2009, 2025)]

sagdp9wv_df = sagdp9wv_df[['GeoName', 'LineCode', 'IndustryClassification', 'Description', 'Unit'] + years]

print('Unnecessary columns dropped successfully!')

# Data check
print('Data shape:', sagdp9wv_df.shape)
sagdp9wv_df.info()
```

```python
# Drop completely empty rows
sagdp9wv_df = sagdp9wv_df.dropna(how='all', axis=0)

print('Completely empty rows dropped successfully!')

print('Data shape:', sagdp9wv_df.shape)
```

```python
# Melt df from wide -> long
# Columns that identify each row
id_vars = ['GeoName', 'LineCode', 'IndustryClassification', 'Description', 'Unit']

# Columns to turn into rows
value_vars = [str(y) for y in range(2009, 2025)]

sagdp9wv_long = pd.melt(
    sagdp9wv_df,
    id_vars=id_vars,
    value_vars=value_vars,
    var_name='Year',
    value_name='GDPReal'
)

print('Data melted successfully!')

print('Data shape:', sagdp9wv_long.shape)
print(sagdp9wv_long.info())
```

```python
# Convert LineCode and Year from float64 to Int64
sagdp9wv_long['LineCode'] = sagdp9wv_long['LineCode'].astype('Int64')

# Convert Year from object to int
sagdp9wv_long['Year'] = sagdp9wv_long['Year'].astype('Int64')

print('Data types converted successfully!')

# Data check
print('Data check:')
print(sagdp9wv_long.info())
sagdp9wv_long.head()
```

```python
# Select the columns we need to map industries
cols_to_check = ['LineCode', 'IndustryClassification', 'Description']

# Drop duplicates so we only see each industry once
bea_industries = sagdp9wv_long[cols_to_check].drop_duplicates()

# Sort by LineCode to see top-level sectors first
bea_industries = bea_industries.sort_values('LineCode')

# Show the rows
print(bea_industries.head(92).to_markdown(index=False))
```

```python
# Strip whitespace from all text columns
text_cols = sagdp9wv_long.select_dtypes(include='object').columns

for col in text_cols:
    sagdp9wv_long[col] = sagdp9wv_long[col].str.strip()

print('Whitespace stripped successfully!')
```

```python
# Map BEA industries to top-level industries
bea_mapping = {
    # Keep top-level total
    'All industry total': 'Total Nonfarm',

    # Agriculture / Mining / Natural resources
    'Farms': 'Mining and Logging',
    'Forestry, fishing, and related activities': 'Mining and Logging',
    'Oil and gas extraction': 'Mining and Logging',
    'Mining (except oil and gas)': 'Mining and Logging',
    'Support activities for mining': 'Mining and Logging',

    # Utilities
    'Utilities': 'Trade, Transportation, and Utilities',

    # Construction
    'Construction': 'Construction',

    # Manufacturing
    'Manufacturing': 'Manufacturing',

    # Trade / Transportation / Utilities
    'Wholesale trade': 'Trade, Transportation, and Utilities',
    'Retail trade': 'Trade, Transportation, and Utilities',
    'Transportation and warehousing': 'Trade, Transportation, and Utilities',

    # Information
    'Information': 'Information',

    # Finance, Insurance, Real Estate, Rental and Leasing
    'Finance and insurance': 'Finance and Insurance',
    'Real estate': 'Real Estate and Rental and Leasing',
    'Rental and leasing services and lessors of nonfinancial intangible assets': 'Real Estate and Rental and Leasing',

    # Professional and Business Services
    'Professional, scientific, and technical services': 'Professional and Business Services',
    'Management of companies and enterprises': 'Professional and Business Services',
    'Administrative and support and waste management and remediation services': 'Professional and Business Services',

    # Education / Health
    'Educational services': 'Private Educational Services',
    'Health care and social assistance': 'Health Care and Social Assistance',

    # Arts / Leisure / Hospitality
    'Arts, entertainment, and recreation': 'Leisure and Hospitality',
    'Accommodation and food services': 'Leisure and Hospitality',

    # Other Services
    'Other services (except government and government enterprises)': 'Other Services',

    # Government
    'Government and government enterprises': 'Government',
}

# Create a new column for industries using the mapping
sagdp9wv_long['IndustryName'] = sagdp9wv_long['Description'].map(bea_mapping)

# Check for any unmapped industries
unmapped = sagdp9wv_long[sagdp9wv_long['IndustryName'].isna()]['Description'].unique()
print('Unmapped BEA industries (if any):', unmapped)

# Drop rows that are parent lines or aggregates that are not needed
sagdp9wv_long = sagdp9wv_long.dropna(subset=['IndustryName'])

# Check the result
sagdp9wv_long.head(20)
```

```python
# Compare totals to check differences (<3% is fine)
# Extract official totals
official_totals = sagdp9wv_long[sagdp9wv_long['Description'] == 'All industry total'][['Year', 'GDPReal']].copy()
official_totals = official_totals.rename(columns={'GDPReal': 'OfficialTotal'})

# Exclude 'All industry total' from mapping
mapped_df = sagdp9wv_long[sagdp9wv_long['Description'] != 'All industry total']

# Sum GDPReal by year for mapped industries
mapped_sums = mapped_df.groupby('Year')['GDPReal'].sum().reset_index()
mapped_sums = mapped_sums.rename(columns={'GDPReal': 'MappedTotal'})

# Merge with official totals
compare_totals = pd.merge(official_totals, mapped_sums, on='Year')
compare_totals['Difference'] = compare_totals['MappedTotal'] - compare_totals['OfficialTotal']
compare_totals['PercentDiff'] = compare_totals['Difference'] / compare_totals['OfficialTotal'] * 100

print(compare_totals.head(20).to_markdown(index=False))
```

```python
# Pull out All industry total for reference
totals = sagdp9wv_long[
    sagdp9wv_long['Description'] == 'All industry total'
][['Year', 'Description', 'GDPReal']].copy()

# Rename for consistency
totals.rename(columns={'Description': 'IndustryName'}, inplace=True)

# Remove All industry total from subsectors before grouping
subsectors = sagdp9wv_long[sagdp9wv_long['Description'] != 'All industry total'].copy()
subsectors = subsectors.dropna(subset=['IndustryName'])

# Sum subsectors by year and mapped industry
grouped = subsectors.groupby(
    ['Year', 'IndustryName'],
    as_index=False
)['GDPReal'].sum()

# Add Total Nonfarm to grouped table
total_nonfarm = totals.copy()
total_nonfarm['IndustryName'] = 'Total Nonfarm'
sagdp9wv_grouped = pd.concat([grouped, total_nonfarm[['Year','IndustryName','GDPReal']]], ignore_index=True)

# Data check
print('Data check:')
print('Data shape:', sagdp9wv_grouped.shape)
sagdp9wv_grouped.head()
```

```python
# Round gdp values
sagdp9wv_grouped['GDPReal'] = sagdp9wv_grouped['GDPReal'].round(1)

# Sort by Year and IndustryName
sagdp9wv_grouped = sagdp9wv_grouped.sort_values(
    ['Year', 'IndustryName']
)

# Data check
print('Data check:')
print('Data shape:', sagdp9wv_grouped.shape)
sagdp9wv_grouped.head()
```

```python
# Exclude Total Nonfarm from industry share calculation
top_level_rows = sagdp9wv_grouped[sagdp9wv_grouped['IndustryName'] != 'Total Nonfarm'].copy()

# Compute total GDP per year from top-level industries
total_gdp_per_year = top_level_rows.groupby('Year')['GDPReal'].sum().reset_index()
total_gdp_per_year.rename(columns={'GDPReal': 'TotalGDP'}, inplace=True)

# Merge TotalGDP back into top-level rows
sagdp9wv_clean = top_level_rows.merge(total_gdp_per_year, on='Year', how='left')

# Compute industry shares
sagdp9wv_clean['IndustryShare'] = sagdp9wv_clean['GDPReal'] / sagdp9wv_clean['TotalGDP']
sagdp9wv_clean['IndustrySharePct'] = sagdp9wv_clean['IndustryShare'] * 100

# Round GDP and shares for readability
sagdp9wv_clean['GDPReal'] = sagdp9wv_clean['GDPReal'].round(1)
sagdp9wv_clean['IndustryShare'] = sagdp9wv_clean['IndustryShare'].round(6)
sagdp9wv_clean['IndustrySharePct'] = sagdp9wv_clean['IndustrySharePct'].round(2)

# Check the result
print('Data check:')
print('Data shape:', sagdp9wv_clean.shape)
sagdp9wv_clean.head()
```

```python
# Sum GDPReal for 2009 (excluding Total Nonfarm)
gdp_2009 = sagdp9wv_clean[sagdp9wv_clean['Year'] == 2009]['GDPReal'].sum()
print(f"Sum of top-level industries for 2009: {gdp_2009}")
```

```python
# Check for duplicates
print('Number of duplicate rows:', sagdp9wv_clean.duplicated().sum())

# Reset index
sagdp9wv_final = sagdp9wv_clean.reset_index(drop=True)

print('Index reset successfully!')
```

```python
# Save df to csv
sagdp9wv_final.to_csv('../../../workspace/nf-insights-project/data_clean/sagdp9wv_clean.csv', index=False)

print('CSV saved successfully!')
```
