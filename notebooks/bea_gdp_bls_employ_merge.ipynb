BEA SAGDP9_WV + BLS Employment dataset merging.

```python
# Import libraries
import pandas as pd

print('Libraries imported successfully')
```

```python
# Read in data
sagdp9wv_df = pd.read_csv('../../../workspace/nf-insights-project/data_clean/sagdp9wv_clean.csv', parse_dates=['Year'], low_memory=False)
blsemp_df = pd.read_csv('../../../workspace/nf-insights-project/data_clean/bls_employ_clean.csv', parse_dates=['Year'], low_memory=False)

print('CSVs read successfully!')

# Data check
print('Data check:')
print(sagdp9wv_df.head())
print(blsemp_df.head())
```

```python
# Data check
print('Data check:')
print(sagdp9wv_df.info())
print(blsemp_df.info())
```

```python
# Check for matching industries
print(set(sagdp9wv_df['Industry']) - set(blsemp_df['Industry']))
print(set(blsemp_df['Industry']) - set(sagdp9wv_df['Industry']))
```

```python
# Filter for years 2009-2024 before merging
sagdp9wv_df = sagdp9wv_df[(sagdp9wv_df['Year'].dt.year >= 2009) & 
                                (sagdp9wv_df['Year'].dt.year <= 2024)]
blsemp_df = blsemp_df[(blsemp_df['Year'].dt.year >= 2009) & 
                            (blsemp_df['Year'].dt.year <= 2024)]

print('Filtered successfully!')

# Data check
print('Data check:')
print(sagdp9wv_df.info())
print(blsemp_df.info())
```

```python
# Merge dataframes
beagdp_blsemp = pd.merge(sagdp9wv_df, blsemp_df, how='inner', on=['Year', 'Industry'], suffixes=('_BEA', '_BLS'))

print('Merged successfully!')

# Reset index
beagdp_blsemp = beagdp_blsemp.reset_index(drop=True)

print('Index reset successfully!')  

# Data check
print('Data check:')
print('Data shape:', beagdp_blsemp.shape)
print(beagdp_blsemp.info())
beagdp_blsemp.head()
```

```python
# Round for readability (2 decimals)
pct_cols = [
    'Industry_Share',
    'Industry_Share_Pct_BEA',
    'Industry_Share_Pct_BLS',
    'Industry_GDP_Growth',
    'State_GDP_Growth',
    'Prior_Share',
    'Growth_Contribution',
    'Growth_3yr_Avg',
    'YoY_Growth',
    'YoY_Growth_3yr_Avg',
    'Top3_Share_Pct'
]

beagdp_blsemp[pct_cols] = beagdp_blsemp[pct_cols].round(2)

print('Rounded to 2 decimals successfully!')

# Round for readability (1 decimal)
conc_cols = ['HHI','Industry_Concentration','Growth_Volatility']

beagdp_blsemp[conc_cols] = beagdp_blsemp[conc_cols].round(1)

print('Rounded to 1 decimal successfully!')

# Convert to int
beagdp_blsemp['Employment'] = beagdp_blsemp['Employment'].astype(int)
beagdp_blsemp['Total_Employment_Year'] = beagdp_blsemp['Total_Employment_Year'].astype(int)
beagdp_blsemp['Yearly_Change'] = beagdp_blsemp['Yearly_Change'].astype(int)

print('Converted to int successfully!')

# Data check
print('Data check:')
print(beagdp_blsemp.info())
beagdp_blsemp.head()
```

```python
# Calculate GDP per worker (productivity)
beagdp_blsemp['GDP_per_Worker'] = (beagdp_blsemp['Real_GDP'] * 1_000_000) / beagdp_blsemp['Employment']

print('GDP per worker calculated successfully!')

# Round for readability
beagdp_blsemp['GDP_per_Worker'] = beagdp_blsemp['GDP_per_Worker'].round(2)

print('Values rounded successfully!')

# Convert to int
beagdp_blsemp['GDP_per_Worker'] = beagdp_blsemp['GDP_per_Worker'].astype(int)

print('Converted to int successfully!')

# Data check
print('Data check:')
print(beagdp_blsemp.info())
beagdp_blsemp.head()
```

```python
# Calculate output vs employment gap
beagdp_blsemp['Output_Gap'] = beagdp_blsemp['Industry_Share_Pct_BEA'] - beagdp_blsemp['Industry_Share_Pct_BLS']

print('Output vs employment gap calculated successfully!')

# Round for readability
beagdp_blsemp['Output_Gap'] = beagdp_blsemp['Output_Gap'].round(2)

print('Values rounded successfully!')

# Data check
print('Data check:')
print(beagdp_blsemp.info())
beagdp_blsemp.head()
```

```python
# Calculate growth contribution vs employment growth to see misalignment
beagdp_blsemp['Employment_vs_Growth_Alignment'] = beagdp_blsemp['YoY_Growth'] * beagdp_blsemp['Growth_Contribution']

print('Growth contribution vs employment growth misalignment calculated successfully!')

# Round for readability
beagdp_blsemp['Employment_vs_Growth_Alignment'] = beagdp_blsemp['Employment_vs_Growth_Alignment'].round(2)

print('Values rounded successfully!')

# Data check
print('Data check:')
print(beagdp_blsemp.info())
beagdp_blsemp.head()
```

```python
# Reset index
beagdp_blsemp_clean = beagdp_blsemp.reset_index(drop=True)

print('Index reset successfully!')

# Save to CSV
beagdp_blsemp_clean.to_csv('../../../workspace/nf-insights-project/data_clean/beagdp_blsemp_merge_clean.csv', index=False)

print('CSV saved successfully!')
```
