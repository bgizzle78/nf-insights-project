BEA SAGDP2 US data cleaning and exploration.

```python
# Import libraries
import pandas as pd

print('Libraries imported successfully!')
```

```python
# Read in data
sagdp2us_df = pd.read_csv('../../../workspace/nf-insights-project/data_raw/BEA GDP Data/SAGDP2_US_1997_2024.csv', low_memory=False)

print('CSV read successfully!')

sagdp2us_df.head()
```

```python
# Data check
print('DF shape: ', sagdp2us_df.shape)
print(sagdp2us_df.info())
sagdp2us_df.columns
```

```python
# Drop unnecessary columns
years = [str(y) for y in range(2009, 2025)]

sagdp2us_df = sagdp2us_df[['GeoName', 'LineCode', 'IndustryClassification', 'Description', 'Unit'] + years]

print('Dropped unnecessary columns successfully!')

# Data check
print('DF shape: ', sagdp2us_df.shape)
sagdp2us_df.info()
```

```python
# Drop completely empty rows
sagdp2us_df = sagdp2us_df.dropna(how='all', axis=0)

print('Dropped completely empty rows successfully!')

print('DF shape: ', sagdp2us_df.shape)
```

```python
# Melt df from wide -> long
# Columns that identify each row
id_vars = ['GeoName', 'LineCode', 'IndustryClassification', 'Description', 'Unit']

# Columns to turn into rows
value_vars = [str(y) for y in range(2009, 2025)]

sagdp2us_long = pd.melt(
    sagdp2us_df,
    id_vars=id_vars,
    value_vars=value_vars,
    var_name='Year',
    value_name='GDPNominal'
)

print('Melted successfully!')

print('DF shape: ', sagdp2us_long.shape)
print(sagdp2us_long.info())
```

```python
# Convert LineCode from float64 to int
sagdp2us_long['LineCode'] = sagdp2us_long['LineCode'].astype(int)

# Convert Year from object to int
sagdp2us_long['Year'] = sagdp2us_long['Year'].astype(int)

print('LineCode and Year converted successfully!')

# Data check
print('Data check:')
print('DF shape: ', sagdp2us_long.shape)
```

```python
# Strip whitespace from all text columns
text_cols = sagdp2us_long.select_dtypes(include='object').columns

for col in text_cols:
    sagdp2us_long[col] = sagdp2us_long[col].str.strip()

print('Whitespace stripped successfully!')

# Reset index
sagdp2us_long = sagdp2us_long.reset_index(drop=True)

print('Index reset successfully!')
```

```python
# Check for duplicates
print('Number of duplicate rows:', sagdp2us_long.duplicated().sum())
```

```python
# Save df to csv
sagdp2us_long.to_csv('../../../workspace/nf-insights-project/data_clean/sagdp2_us_clean.csv', index=False)

print('CSV saved successfully!')
```
