WVSOS Business Registrations data cleaning and exploration.

```python
# Import libraries
import pandas as pd
```

```python
# Read in wvsos business registration data csv file
wvsos_df = pd.read_csv('../../../workspace/NF Insights Project/data_raw/WVSOS Business Registrations - WVSOS.csv', low_memory=False)
print('CSV read successfully!')

wvsos_df.head()
```

```python
wvsos_df.info()
```

```python
wvsos_df.columns
```

```python
# Convert date columns to datetime
date_columns = [
    'Filing Date',
    'Termination Date'
]

wvsos_df[date_columns] = wvsos_df[date_columns].apply(
    pd.to_datetime,
    errors='coerce'
)

wvsos_df.head()
```

```python
# Convert NAICS to numeric
wvsos_df['NAICS'] = wvsos_df['NAICS'].astype('Int64')

wvsos_df.head()
```

```python
# Filter for years 2009-2025
wvsos_df = wvsos_df[
    (wvsos_df['Filing Date'] >= '2009-01-01') &
    (wvsos_df['Filing Date'] <= '2025-12-31')
]
```

```python
# Drop unneeded columns
wvsos_df = wvsos_df.drop(columns=[
    'ID',
    'URL',
    'Effective Date',
    'Established Date',
    'Charter',
    'Class',
    'Termination Reason',
    'Sec Type',   
    'Addresses Type',
    'Addresses Address'
])

wvsos_df.head()
```

```python
# Extract year and month from Filing Date
wvsos_df['Filing Year'] = wvsos_df['Filing Date'].dt.year.astype('Int64')
wvsos_df['Filing Month'] = wvsos_df['Filing Date'].dt.month.astype('Int64')

# Extract year and month from Termination Date
wvsos_df['Termination Year'] = wvsos_df['Termination Date'].dt.year.astype('Int64')
wvsos_df['Termination Month'] = wvsos_df['Termination Date'].dt.month.astype('Int64')

wvsos_df.head()
```

```python
# Define desired column order
cols_order = [
    'Organization Name', 'Org Type', 
    'Filing Date', 'Filing Year', 'Filing Month',
    'Termination Date', 'Termination Year', 'Termination Month',
    'Business Purpose', 'Main Industry', 'Sub-Industry', 'Business Type',
    'NAICS', 'Charter County', 'Charter State'
]

# Reorder the DataFrame
wvsos_df  = wvsos_df[cols_order]

wvsos_df.head()
```

```python
# Save cleaned data
wvsos_df.to_csv('../../../workspace/NF Insights Project/data_clean/wvsos_business_registrations_clean.csv', index=False)

print('CSV saved successfully!')
```
