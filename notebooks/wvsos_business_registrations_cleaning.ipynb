WVSOS Business Registrations data cleaning and exploration.

```python
# Import libraries
import pandas as pd

print('Libraries imported successfully!')
```

```python
# Read in wvsos business registration data
wvsos_df = pd.read_csv('../../../workspace/nf-insights-project/data_raw/WVSOS Business Registrations Complete Data -Needs Cleaning.csv', low_memory=False)

print('CSV read successfully!')

# Data check
print('Data check:')
print('DF shape:', wvsos_df.shape)
print(wvsos_df.info())
wvsos_df.head()
```

```python
# Convert date columns to datetime
date_columns = [
    'Filing Date',
    'Termination Date'
]

wvsos_df[date_columns] = wvsos_df[date_columns].apply(
    pd.to_datetime,
    errors='coerce'
)

print('Date columns converted successfully!')

# Convert NAICS to numeric
wvsos_df['NAICS'] = wvsos_df['NAICS'].astype('Int64')

print('NAICS converted to numeric successfully!')

# Data check
print('Data check:')
wvsos_df.info()
```

```python
# Filter for years 2008-2025
wvsos_df = wvsos_df[
    (wvsos_df['Filing Date'] >= '2008-01-01') &
    (wvsos_df['Filing Date'] <= '2025-12-31')
]

print('Data filtered for years 2008-2025 successfully!')

# Data check
print('Data check:')
print('DF shape:', wvsos_df.shape)
wvsos_df.info()
```

```python
# Drop unneeded columns
wvsos_df = wvsos_df.drop(columns=[
    'URL',
    'Effective Date',
    'Established Date',
    'Charter',
    'Class',
    'Termination Reason',
    'Sec Type', 
    'Charter County',
    'At Will Term',
    'At Will Term Years',
    'Authorized Shares',
    'Capital Stock',
    'Control Number',
    'Excess Acres',
    'Member Managed',
    'Par Value',
    'Young Entrepreneur',  
    'Addresses Type',
    'Addresses Address',
    'Officers Type',
    'Officers Name/Address',
    'DBA Name',
    'DBA Description',
    'DBA Effective Date',
    'DBA Termination Date',
    'Amendments',
    'Amendment Date',
    'Mergers Merged',
    'Mergers Merged State',
    'Mergers Merger Date',
    'Mergers Survived',
    'Mergers Survived State',
    'Dissolutions Action Pending',
    'Dissolutions Approval Request Date',
    'Dissolutions Articles Date',
    'Dissolutions Employment Security Approval Date',
    'Dissolutions Tax Approval Date',
    'Dissolutions Workers Compensation Approval Date',
    'Subsidiaries Address',
    'Subsidiaries Name',
    'Last Annual Report',
    'Unnamed: 48',
    'Education Level',
    'Education Status',
    'Field',
    'School'
])

print('Unneeded columns dropped successfully!')        

# Data check
print('Data check:')
print('Data shape:', wvsos_df.shape)
wvsos_df.info()
```

```python
# Check for duplicates
print('Number of duplicates:', wvsos_df.duplicated().sum())
```

```python
# View the duplicate rows
duplicates = wvsos_df[wvsos_df.duplicated()]

print('Duplicate rows pulled successfully!')

# View the duplicate rows
print('Duplicate rows:')
duplicates.head()
```

```python
# Extract year from Filing Date
wvsos_df['Filing Year'] = wvsos_df['Filing Date'].dt.year.astype('Int64')

# Extract year from Termination Date
wvsos_df['Termination Year'] = wvsos_df['Termination Date'].dt.year.astype('Int64')

print('Year columns extracted successfully!')

# Data check
print('Data check:')
print('DF shape:', wvsos_df.shape)
wvsos_df.info()
```

```python
# Strip whitespace from all text columns
text_cols = wvsos_df.select_dtypes(include='object').columns

for col in text_cols:
    wvsos_df[col] = wvsos_df[col].str.strip()

print('Whitespace stripped successfully!')
```

```python
# Unique NAICS codes
print(sorted(wvsos_df['NAICS'].dropna().unique().tolist()))

print('Unique NAICS pulled successfully!')
```

```python
# Create 2-digit NAICS sector for mapping to industry
wvsos_df['NAICS_2'] = (
    wvsos_df['NAICS']
    .astype('Int64')
    .astype(str)
    .str[:2]
)

print('2-digit NAICS created successfully!')

# Data check
print('Data check:')
wvsos_df.info()
```

```python
# Pull unique NAICS_2 codes
print(sorted(wvsos_df['NAICS_2'].dropna().unique().tolist()))

print('Unique NAICS_2 pulled successfully!')
```

```python
# Map WVSOS NAICS industries to BEA/BLS categories
def map_industry(sector):
    if sector in ['21', '11']:
        return 'Mining and Logging'
    elif sector == '23':
        return 'Construction'
    elif sector in ['31', '32', '33']:
        return 'Manufacturing'
    elif sector in ['42', '44', '45', '48', '49', '22']:
        return 'Trade, Transportation, and Utilities'
    elif sector == '51':
        return 'Information'
    elif sector == '52':
        return 'Finance and Insurance'
    elif sector == '53':
        return 'Real Estate and Rental and Leasing'
    elif sector in ['54', '55', '56']:
        return 'Professional and Business Services'
    elif sector == '61':
        return 'Private Education Services'
    elif sector == '62':
        return 'Health Care and Social Assistance'
    elif sector in ['71', '72']:
        return 'Leisure and Hospitality'
    elif sector == '81':
        return 'Other Services'
    elif sector == '92':
        return 'Government'
    else:
        return 'Unmapped'

wvsos_df['Industry'] = wvsos_df['NAICS_2'].apply(map_industry)

print('NAICS mapped successfully!')

# Data check
print('Data check:')
print('DF shape:', wvsos_df.shape)
print(wvsos_df.info())
wvsos_df.head()
```

```python
# Filter out TMO (trademark holder) registrations before metric calculations
wvsos_df = wvsos_df[wvsos_df['Org Type'] != 'TMO | Trademark Holder']

print('TMO (trademark holder) registrations removed successfully!')

# Null termination dates after 2025
wvsos_df.loc[wvsos_df["Termination Year"] > 2025, "Termination Year"] = pd.NA

print('Nulled termination dates after 2025 successfully!')

# Data check
print('Data check:')
print('DF shape:', wvsos_df.shape)
print(wvsos_df.info())
```

```python
# Remove more unneeded columns before starting metrics
wvsos_df = wvsos_df.drop(columns=['Filing Date', 'Termination Date', 'Business Purpose', 'NAICS', 'NAICS_2', 'Charter State'])

print('Unneeded columns removed successfully!')

# Data check
print('Data check:')
print('DF shape:', wvsos_df.shape)
print(wvsos_df.info())
wvsos_df.head()
```

```python
# Count new registrations by year and industry
registrations_per_year_industry = wvsos_df.groupby(['Filing Year', 'Industry']).size()

# Map counts back to wvsos_df
wvsos_df['Yearly Registrations'] = wvsos_df.set_index(['Filing Year', 'Industry']).index.map(registrations_per_year_industry)

print('Yearly registrations added successfully!')

# Fill nulls with 0 and convert to int
wvsos_df['Yearly Registrations'] = wvsos_df['Yearly Registrations'].fillna(0).astype(int)

# Data check
print('Data check:')
print('Data shape:', wvsos_df.shape)
wvsos_df.head()
```

```python
# Count closures by year and industry
closures_per_year_industry = wvsos_df.groupby(['Termination Year', 'Industry']).size()

# Map counts back to wvsos_df
wvsos_df['Yearly Closures'] = wvsos_df.set_index(['Termination Year', 'Industry']).index.map(closures_per_year_industry)

print('Yearly closures added successfully!')

# Fill nulls with 0 and convert to int
wvsos_df['Yearly Closures'] = wvsos_df['Yearly Closures'].fillna(0).astype(int)

# Data check
print('Data check:')
print('Data shape:', wvsos_df.shape)
wvsos_df.head()
```

```python
# Calculate net business changes by year
wvsos_df['Net Business Change'] = wvsos_df['Yearly Registrations'] - wvsos_df['Yearly Closures']

print('Net business change added successfully!')

# Data check
print('Data check:')
print('Data shape:', wvsos_df.shape)
wvsos_df.head()
```

```python
# Groupby total registrations per year
total_registrations_per_year = wvsos_df.groupby('Filing Year').size().to_dict()

# Total registrations per year and industry
registrations_per_year_industry = wvsos_df.groupby(['Filing Year', 'Industry']).size().to_dict()

# Calculate registration share at industry-year level
def compute_share(row):
    key = (row['Filing Year'], row['Industry'])
    industry_total = registrations_per_year_industry.get(key, 0)
    year_total = total_registrations_per_year.get(row['Filing Year'], 1)
    return industry_total / year_total

wvsos_df['Registration Share'] = wvsos_df.apply(compute_share, axis=1)

print('Registration industry share added successfully!')

# Round for readability
wvsos_df['Registration Share'] = wvsos_df['Registration Share'].round(2)

print('Registration industry share rounded successfully!')

# Data check
print('Data check:')
print('Data shape:', wvsos_df.shape)
wvsos_df.head()
```

```python
# Calculate HHI in 0-10000 scale
hhi_per_year = (
    wvsos_df[['Filing Year', 'Industry', 'Registration Share']]
    .drop_duplicates()
    .groupby('Filing Year')['Registration Share']
    .apply(lambda x: (x**2).sum() * 10000)
)

wvsos_df['HHI Business Formation'] = wvsos_df['Filing Year'].map(hhi_per_year)

print('HHI added successfully!')

# Change dtype from float to int
wvsos_df['HHI Business Formation'] = wvsos_df['HHI Business Formation'].astype(int)

print('HHI dtype changed successfully!')

# Data check
print('Data check:')
print('Data shape:', wvsos_df.shape)
wvsos_df.head()
```

```python
# Calculate industry net growth per year
# Sum net change per industry-year
industry_year_sum = wvsos_df.groupby(['Filing Year', 'Industry'], as_index=False)['Net Business Change'].sum()

# Calculate total net change per year
total_per_year = industry_year_sum.groupby('Filing Year')['Net Business Change'].sum()

# Calculate share per industry-year
industry_year_sum['Net Growth Share'] = (
    industry_year_sum['Net Business Change'] / industry_year_sum['Filing Year'].map(total_per_year)
)

# Map the correct share back to every row in the original dataframe
wvsos_df['Net Growth Share'] = wvsos_df.set_index(['Filing Year','Industry']).index.map(
    dict(zip(
        zip(industry_year_sum['Filing Year'], industry_year_sum['Industry']),
        industry_year_sum['Net Growth Share']
    ))
)

print('Net business growth added successfully!')

# Round for readability
wvsos_df['Net Growth Share'] = wvsos_df['Net Growth Share'].round(2)

print('Net business growth rounded successfully!')

# Data check
print('Data check:')
print('Data shape:', wvsos_df.shape)
wvsos_df.head()
```

```python
# Calculate net business growth rate per year
# Sum net change per industry-year
industry_year_net = wvsos_df.groupby(['Filing Year','Industry'], as_index=False)['Net Business Change'].sum()

# Calculate pct_change per industry
industry_year_net['Net Growth Rate'] = industry_year_net.groupby('Industry')['Net Business Change'].pct_change() * 100

# Map back to every row in wvsos_df
wvsos_df['Net Growth Rate'] = wvsos_df.set_index(['Filing Year','Industry']).index.map(
    dict(zip(
        zip(industry_year_net['Filing Year'], industry_year_net['Industry']),
        industry_year_net['Net Growth Rate']
    ))
)

print('Net business growth rate added successfully!')

# Round for readability
wvsos_df['Net Growth Rate'] = wvsos_df['Net Growth Rate'].round(2)

print('Net business growth rate rounded successfully!')

# Data check
print('Data check:')
print('Data shape:', wvsos_df.shape)
wvsos_df.head()
```

```python
# Reset index
wvsos_clean = wvsos_df.reset_index(drop=True)

print('Index reset successfully!')

# Save cleaned data
wvsos_clean.to_csv('../../../workspace/nf-insights-project/data_clean/wvsos_business_registrations_clean.csv', index=False)

print('CSV saved successfully!')
```
