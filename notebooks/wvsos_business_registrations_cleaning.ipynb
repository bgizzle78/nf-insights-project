WVSOS Business Registrations data cleaning and exploration.

```python
# Import libraries
import pandas as pd

print('Libraries imported successfully!')
```

```python
# Read in wvsos business registration data
wvsos_df = pd.read_csv('../../../workspace/nf-insights-project/data_raw/WVSOS Business Registrations Complete Data -Needs Cleaning.csv', low_memory=False)

print('CSV read successfully!')

# Data check
print('Data check:')
print('DF shape:', wvsos_df.shape)
print(wvsos_df.info())
wvsos_df.head()
```

```python
# Convert date columns to datetime
date_columns = [
    'Filing Date',
    'Termination Date'
]

wvsos_df[date_columns] = wvsos_df[date_columns].apply(
    pd.to_datetime,
    errors='coerce'
)

print('Date columns converted successfully!')

# Convert NAICS to numeric
wvsos_df['NAICS'] = wvsos_df['NAICS'].astype('Int64')

print('NAICS converted to numeric successfully!')

# Data check
print('Data check:')
wvsos_df.info()
```

```python
# Filter for years 2008-2025
wvsos_df = wvsos_df[
    (wvsos_df['Filing Date'] >= '2008-01-01') &
    (wvsos_df['Filing Date'] <= '2025-12-31')
]

print('Data filtered for years 2008-2025 successfully!')

# Data check
print('Data check:')
print('DF shape:', wvsos_df.shape)
wvsos_df.info()
```

```python
# Drop unneeded columns
wvsos_df = wvsos_df.drop(columns=[
    'URL',
    'Effective Date',
    'Established Date',
    'Charter',
    'Class',
    'Termination Reason',
    'Sec Type', 
    'Charter County',
    'At Will Term',
    'At Will Term Years',
    'Authorized Shares',
    'Capital Stock',
    'Control Number',
    'Excess Acres',
    'Member Managed',
    'Par Value',
    'Young Entrepreneur',  
    'Addresses Type',
    'Addresses Address',
    'Officers Type',
    'Officers Name/Address',
    'DBA Name',
    'DBA Description',
    'DBA Effective Date',
    'DBA Termination Date',
    'Amendments',
    'Amendment Date',
    'Mergers Merged',
    'Mergers Merged State',
    'Mergers Merger Date',
    'Mergers Survived',
    'Mergers Survived State',
    'Dissolutions Action Pending',
    'Dissolutions Approval Request Date',
    'Dissolutions Articles Date',
    'Dissolutions Employment Security Approval Date',
    'Dissolutions Tax Approval Date',
    'Dissolutions Workers Compensation Approval Date',
    'Subsidiaries Address',
    'Subsidiaries Name',
    'Last Annual Report',
    'Unnamed: 48',
    'Education Level',
    'Education Status',
    'Field',
    'School'
])

print('Unneeded columns dropped successfully!')        

# Data check
print('Data check:')
print('Data shape:', wvsos_df.shape)
wvsos_df.info()
```

```python
# Check for duplicates
print('Number of duplicates:', wvsos_df.duplicated().sum())
```

```python
# View the duplicate rows
duplicates = wvsos_df[wvsos_df.duplicated()]

print('Duplicate rows pulled successfully!')

# View the duplicate rows
print('Duplicate rows:')
duplicates.head()
```

```python
# Extract year from Filing Date
wvsos_df['Filing Year'] = wvsos_df['Filing Date'].dt.year.astype('Int64')
wvsos_df['Filing Month'] = wvsos_df['Filing Date'].dt.month.astype('Int64')

# Extract year from Termination Date
wvsos_df['Termination Year'] = wvsos_df['Termination Date'].dt.year.astype('Int64')
wvsos_df['Termination Month'] = wvsos_df['Termination Date'].dt.month.astype('Int64')

print('Year columns extracted successfully!')

# Data check
print('Data check:')
wvsos_df.info()
```

```python
# Strip whitespace from all text columns
text_cols = wvsos_df.select_dtypes(include='object').columns

for col in text_cols:
    wvsos_df[col] = wvsos_df[col].str.strip()

print('Whitespace stripped successfully!')
```

```python
# Unique NAICS codes
print(sorted(wvsos_df["NAICS"].dropna().unique().tolist()))

print('Unique NAICS pulled successfully!')
```

```python
# Create 2-digit NAICS sector
wvsos_df["NAICS_2"] = (
    wvsos_df["NAICS"]
    .astype("Int64")
    .astype(str)
    .str[:2]
)

print('2-digit NAICS created successfully!')

# Data check
print('Data check:')
wvsos_df.info()
```

```python
sorted(wvsos_df["NAICS_2"].dropna().unique())
```

```python
# Map WVSOS NAICS industries to BEA/BLS categories
def map_industry(sector):
    if sector in ["21", "11"]:
        return "Mining and Logging"
    elif sector == "23":
        return "Construction"
    elif sector in ["31", "32", "33"]:
        return "Manufacturing"
    elif sector in ["42", "44", "45", "48", "49", "22"]:
        return "Trade, Transportation, and Utilities"
    elif sector == "51":
        return "Information"
    elif sector == "52":
        return "Finance and Insurance"
    elif sector == "53":
        return "Real Estate and Rental and Leasing"
    elif sector in ["54", "55", "56"]:
        return "Professional and Business Services"
    elif sector == "61":
        return "Private Education Services"
    elif sector == "62":
        return "Health Care and Social Assistance"
    elif sector in ["71", "72"]:
        return "Leisure and Hospitality"
    elif sector == "81":
        return "Other Services"
    elif sector == "92":
        return "Government"
    else:
        return "Unmapped"

wvsos_df["IndustryMapped"] = wvsos_df["NAICS_2"].apply(map_industry)

print('NAICS mapped successfully!')

# Data check
print('Data check:')
print(wvsos_df.info())
wvsos_df.head()
```

```python
# Filter out TMO (trademark holder) registrations before metric calculations
wvsos_df = wvsos_df[wvsos_df['Org Type'] != 'TMO | Trademark Holder']

print('TMO (trademark holder) registrations removed successfully!')
```

```python
# Count new registrations by year
registrations_per_year = wvsos_df.groupby('Filing Year').size().to_dict()

# Add a new column to wvsos_df with yearly registrations
wvsos_df['Yearly_Registrations'] = wvsos_df['Filing Year'].map(registrations_per_year)

print('Yearly registrations added successfully!')

# Data check
print('Data check:')
print('Data shape:', wvsos_df.shape)
print(wvsos_df.info())
wvsos_df.head()
```

```python
# Count terminated businesses by year
closures_per_year = wvsos_df.groupby('Termination Year').size().to_dict()

# Add a new column to wvsos_df with yearly closures
wvsos_df['Yearly_Closures'] = wvsos_df['Termination Year'].map(closures_per_year)

print('Yearly closures added successfully!')

# Data check
print('Data check:')
print('Data shape:', wvsos_df.shape)
print(wvsos_df.info())
wvsos_df.head()
```
