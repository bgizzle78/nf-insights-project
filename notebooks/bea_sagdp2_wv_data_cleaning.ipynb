BEA SAGDP2 WV data cleaning and exploration.

```python
# Import libraries
import pandas as pd

print('Libraries imported successfully!')
```

```python
# Read in data
sagdp2wv_df = pd.read_csv('../../../workspace/nf-insights-project/data_raw/BEA GDP Data/SAGDP2_WV_1997_2024.csv', low_memory=False)

print('CSV read successfully!')

sagdp2wv_df.head()
```

```python
# Data check
print('Data check:')
print('Data shape:', sagdp2wv_df.shape)
print(sagdp2wv_df.info())
sagdp2wv_df.columns
```

```python
# Drop unnecessary columns
years = [str(y) for y in range(2008, 2025)]

sagdp2wv_df = sagdp2wv_df[['GeoName', 'LineCode', 'IndustryClassification', 'Description', 'Unit'] + years]

print('Unnecessary columns dropped successfully!')

# Data check
print('Data shape:', sagdp2wv_df.shape)
sagdp2wv_df.info()
```

```python
# Drop completely empty rows
sagdp2wv_df = sagdp2wv_df.dropna(how='all', axis=0)

print('Completely empty rows dropped successfully!')

print('Data shape:', sagdp2wv_df.shape)
```

```python
# Melt df from wide -> long
# Columns that identify each row
id_vars = ['GeoName', 'LineCode', 'IndustryClassification', 'Description', 'Unit']

# Columns to turn into rows
value_vars = [str(y) for y in range(2008, 2025)]

sagdp2wv_long = pd.melt(
    sagdp2wv_df,
    id_vars=id_vars,
    value_vars=value_vars,
    var_name='Year',
    value_name='GDPNominal'
)

print('Data melted successfully!')

print('Data shape:', sagdp2wv_long.shape)
print(sagdp2wv_long.info())
```

```python
# Convert LineCode from float64 to int
sagdp2wv_long['LineCode'] = sagdp2wv_long['LineCode'].astype('Int64')

# Convert Year from object to int
sagdp2wv_long['Year'] = sagdp2wv_long['Year'].astype('Int64')

print('Data types converted successfully!')

# Data check
print('Data check:')
print(sagdp2wv_long.info())
sagdp2wv_long.head()
```

```python
# Select the columns we need to map industries
cols_to_check = ['LineCode', 'IndustryClassification', 'Description']

# Drop duplicates so we only see each industry once
bea_industries = sagdp2wv_long[cols_to_check].drop_duplicates()

# Sort by LineCode to see top-level sectors first
bea_industries = bea_industries.sort_values('LineCode')

# Show the rows
print(bea_industries.head(92).to_markdown(index=False))
```

```python
# Strip whitespace from all text columns
text_cols = sagdp2wv_long.select_dtypes(include='object').columns

for col in text_cols:
    sagdp2wv_long[col] = sagdp2wv_long[col].str.strip()

print('Whitespace stripped successfully!')
```

```python
# Map BEA industries to top-level industries
bea_mapping = {
    # Keep top-level total
    'All industry total': 'Total Nonfarm',

    # Agriculture / Mining / Natural resources
    'Farms': 'Mining and Logging',
    'Forestry, fishing, and related activities': 'Mining and Logging',
    'Oil and gas extraction': 'Mining and Logging',
    'Mining (except oil and gas)': 'Mining and Logging',
    'Support activities for mining': 'Mining and Logging',

    # Utilities
    'Utilities': 'Trade, Transportation, and Utilities',

    # Construction
    'Construction': 'Construction',

    # Manufacturing
    'Manufacturing': 'Manufacturing',

    # Trade / Transportation / Utilities
    'Wholesale trade': 'Trade, Transportation, and Utilities',
    'Retail trade': 'Trade, Transportation, and Utilities',
    'Transportation and warehousing': 'Trade, Transportation, and Utilities',

    # Information
    'Information': 'Information',

    # Finance, Insurance, Real Estate, Rental and Leasing
    'Finance and insurance': 'Finance and Insurance',
    'Real estate': 'Real Estate and Rental and Leasing',
    'Rental and leasing services and lessors of nonfinancial intangible assets': 'Real Estate and Rental and Leasing',

    # Professional and Business Services
    'Professional, scientific, and technical services': 'Professional and Business Services',
    'Management of companies and enterprises': 'Professional and Business Services',
    'Administrative and support and waste management and remediation services': 'Professional and Business Services',

    # Education / Health
    'Educational services': 'Private Educational Services',
    'Health care and social assistance': 'Health Care and Social Assistance',

    # Arts / Leisure / Hospitality
    'Arts, entertainment, and recreation': 'Leisure and Hospitality',
    'Accommodation and food services': 'Leisure and Hospitality',

    # Other Services
    'Other services (except government and government enterprises)': 'Other Services',

    # Government
    'Government and government enterprises': 'Government',
}

# Create a new column for industries using the mapping
sagdp2wv_long['IndustryName'] = sagdp2wv_long['Description'].map(bea_mapping)

# Check for any unmapped industries
unmapped = sagdp2wv_long[sagdp2wv_long['IndustryName'].isna()]['Description'].unique()
print('Unmapped BEA industries (if any):', unmapped)

# Drop rows that are parent lines or aggregates that are not needed
sagdp2wv_long = sagdp2wv_long.dropna(subset=['IndustryName'])

# Check the result
sagdp2wv_long.head(20)
```

```python
# Compare totals to check differences (<3% is fine)
# Extract official totals
official_totals = sagdp2wv_long[sagdp2wv_long['Description'] == 'All industry total'][['Year', 'GDPNominal']].copy()
official_totals = official_totals.rename(columns={'GDPNominal': 'OfficialTotal'})

# Exclude 'All industry total' from mapping
mapped_df = sagdp2wv_long[sagdp2wv_long['Description'] != 'All industry total']

# Sum GDPReal by year for mapped industries
mapped_sums = mapped_df.groupby('Year')['GDPNominal'].sum().reset_index()
mapped_sums = mapped_sums.rename(columns={'GDPNominal': 'MappedTotal'})

# Merge with official totals
compare_totals = pd.merge(official_totals, mapped_sums, on='Year')
compare_totals['Difference'] = compare_totals['MappedTotal'] - compare_totals['OfficialTotal']
compare_totals['PercentDiff'] = compare_totals['Difference'] / compare_totals['OfficialTotal'] * 100

print(compare_totals.head(20).to_markdown(index=False))
```

```python
# Pull out All industry total for reference
totals = sagdp2wv_long[
    sagdp2wv_long['Description'] == 'All industry total'
][['Year', 'Description', 'GDPNominal']].copy()

# Rename for consistency
totals.rename(columns={'Description': 'IndustryName'}, inplace=True)

# Remove All industry total from subsectors before grouping
subsectors = sagdp2wv_long[sagdp2wv_long['Description'] != 'All industry total'].copy()
subsectors = subsectors.dropna(subset=['IndustryName'])

# Sum subsectors by year and mapped industry
grouped = subsectors.groupby(
    ['Year', 'IndustryName'],
    as_index=False
)['GDPNominal'].sum()

# Add Total Nonfarm to grouped table
total_nonfarm = totals.copy()
total_nonfarm['IndustryName'] = 'Total Nonfarm'
sagdp2wv_grouped = pd.concat([grouped, total_nonfarm[['Year','IndustryName','GDPNominal']]], ignore_index=True)

# Data check
print('Data check:')
print('Data shape:', sagdp2wv_grouped.shape)
sagdp2wv_grouped.head()
```

```python
# Round gdp values
sagdp2wv_grouped['GDPNominal'] = sagdp2wv_grouped['GDPNominal'].round(1)

# Sort by Year and IndustryName
sagdp2wv_grouped = sagdp2wv_grouped.sort_values(
    ['Year', 'IndustryName']
)

# Data check
print('Data check:')
print('Data shape:', sagdp2wv_grouped.shape)
sagdp2wv_grouped.head()
```

```python
# Exclude Total Nonfarm from industry share calculation
top_level_rows = sagdp2wv_grouped[sagdp2wv_grouped['IndustryName'] != 'Total Nonfarm'].copy()

# Compute total GDP per year from top-level industries
total_gdp_per_year = top_level_rows.groupby('Year')['GDPNominal'].sum().reset_index()
total_gdp_per_year.rename(columns={'GDPNominal': 'TotalGDP'}, inplace=True)

# Merge TotalGDP back into top-level rows
sagdp2wv_clean = top_level_rows.merge(total_gdp_per_year, on='Year', how='left')

# Compute industry shares
sagdp2wv_clean['IndustryShare'] = sagdp2wv_clean['GDPNominal'] / sagdp2wv_clean['TotalGDP']
sagdp2wv_clean['IndustrySharePct'] = sagdp2wv_clean['IndustryShare'] * 100

# Round GDP and shares for readability
sagdp2wv_clean['GDPNominal'] = sagdp2wv_clean['GDPNominal'].round(1)
sagdp2wv_clean['IndustryShare'] = sagdp2wv_clean['IndustryShare'].round(6)
sagdp2wv_clean['IndustrySharePct'] = sagdp2wv_clean['IndustrySharePct'].round(2)

# Check the result
print('Data check:')
print('Data shape:', sagdp2wv_clean.shape)
sagdp2wv_clean.head()
```

```python
# Sum GDPReal for each year across top-level industries only (exclude Total Nonfarm)
yearly_totals = sagdp2wv_clean.groupby('Year')['GDPNominal'].sum().reset_index()

print(yearly_totals)
```

```python
# Calculate year-over-year industry GDP growth
sagdp2wv_clean['IndustryGDPGrowth'] = sagdp2wv_clean.groupby('IndustryName')['GDPNominal'].pct_change() * 100

# Round industry growth for readability
sagdp2wv_clean['IndustryGDPGrowth'] = sagdp2wv_clean['IndustryGDPGrowth'].round(2)

# Data check
print('Data check:')
print('Data shape:', sagdp2wv_clean.shape)
sagdp2wv_clean.head()
```

```python
# Compute total GDP per year (unique yearly values)
state_growth = (
    sagdp2wv_clean[['Year', 'TotalGDP']]
    .drop_duplicates()
    .sort_values('Year')
)

# Calculate year-over-year growth
state_growth['StateGDPGrowth'] = (
    state_growth['TotalGDP'].pct_change() * 100
)

# Round
state_growth['StateGDPGrowth'] = state_growth['StateGDPGrowth'].round(2)

# Merge back into main dataframe
sagdp2wv_clean = sagdp2wv_clean.merge(
    state_growth[['Year', 'StateGDPGrowth']],
    on='Year',
    how='left'
)

# Data check
print('Data check:')
print('Data shape:', sagdp2wv_clean.shape)
sagdp2wv_clean.head()
```

```python
# Calculate contribution to growth
# Shift industry share back one year
sagdp2wv_clean['PriorShare'] = (
    sagdp2wv_clean
    .groupby('IndustryName')['IndustryShare']
    .shift(1)
)

# Compute contribution to state growth
sagdp2wv_clean['GrowthContribution'] = (
    sagdp2wv_clean['PriorShare'] *
    sagdp2wv_clean['IndustryGDPGrowth']
)

# Round for readability
sagdp2wv_clean['GrowthContribution'] = sagdp2wv_clean['GrowthContribution'].round(2)

# Data check
print('Data check:')
print('Data shape:', sagdp2wv_clean.shape)
sagdp2wv_clean.head()
```

```python
# Calculate industry concentration (HHI) in 0-10,000 scale
sagdp2wv_clean['IndustryConcentration'] = (sagdp2wv_clean['IndustryShare']*100) ** 2

# Compute HHI per year
hhi_by_year = (
    sagdp2wv_clean
    .groupby('Year')['IndustryConcentration']
    .sum()
    .reset_index()
)

# Round for readability
sagdp2wv_clean['IndustryConcentration'] = sagdp2wv_clean['IndustryConcentration'].round(1)

# Check the result
print('Data check:')
print('Data shape:', sagdp2wv_clean.shape)
sagdp2wv_clean.head()
```

```python
# Compute 3-year rolling average growth per industry
sagdp2wv_clean['Growth3yrAvg'] = (
    sagdp2wv_clean
    .groupby('IndustryName')['IndustryGDPGrowth']
    .rolling(3)
    .mean()
    .reset_index(level=0, drop=True)
)

# Round for readability
sagdp2wv_clean['Growth3yrAvg'] = sagdp2wv_clean['Growth3yrAvg'].round(2)

# Data check
print('Data check:')
print('Data shape:', sagdp2wv_clean.shape)
sagdp2wv_clean.head()
```

```python
# Calculate volatility per industry
sagdp2wv_clean['GrowthVolatility'] = (
    sagdp2wv_clean
    .groupby('IndustryName')['IndustryGDPGrowth']
    .transform('std')
)

sagdp2wv_clean['GrowthVolatility'] = sagdp2wv_clean['GrowthVolatility'].round(2)

# Data check
print('Data check:')
print('Data shape:', sagdp2wv_clean.shape)
sagdp2wv_clean.head()
```

```python
# Calculate industry rank by size (structural view)
sagdp2wv_clean['IndustryRank'] = (
    sagdp2wv_clean
    .groupby('Year')['GDPNominal']
    .rank(ascending=False, method='min')
)

# Convert rank to integer
sagdp2wv_clean['IndustryRank'] = sagdp2wv_clean['IndustryRank'].astype(int)

# Data check
print('Data check:')
print('Data shape:', sagdp2wv_clean.shape)
sagdp2wv_clean.head()
```

```python
# Check for duplicates
print('Number of duplicate rows:', sagdp2wv_clean.duplicated().sum())

# Reset index
sagdp2wv_final = sagdp2wv_clean.reset_index(drop=True)

print('Index reset successfully!')
```

```python
# Save df to csv
sagdp2wv_final.to_csv('../../../workspace/nf-insights-project/data_clean/sagdp2wv_clean.csv', index=False)

print('CSV saved successfully!')
```
