BEA SAGDP1 All Areas data cleaning and exploration.

```python
# Import libraries
import pandas as pd

print('Libraries imported successfully!')
```

```python
# Read in data
sagdp1_df = pd.read_csv('../../../workspace/nf-insights-project/data_raw/BEA GDP Data/SAGDP1__ALL_AREAS_1997_2024.csv', low_memory=False)

print('CSV read successfully!')

sagdp1_df.head()
```

```python
# Data check
print('Data check:')
print('DF shape:', sagdp1_df.shape)
print(sagdp1_df.info())
sagdp1_df.columns
```

```python
# Drop unnecessary columns
years = [str(y) for y in range(2009, 2025)]

sagdp1_df = sagdp1_df[['GeoName', 'LineCode', 'IndustryClassification', 'Description', 'Unit'] + years] 

print('Dropped unnecessary columns successfully!')

# Data check
print('DF shape:' ,sagdp1_df.shape)
sagdp1_df.info()
```

```python
# Drop completely empty rows
sagdp1_df = sagdp1_df.dropna(how='all', axis=0)

print('Dropped completely empty rows successfully!')

print('DF shape:', sagdp1_df.shape)
```

```python
# Keep only WV and US rows
sagdp1_df = sagdp1_df[sagdp1_df['GeoName'].isin(['West Virginia', 'United States'])]

print('Kept only WV and US rows successfully!')

print('DF shape:', sagdp1_df.shape)

sagdp1_df['GeoName'].unique()
```

```python
# Melt df from wide -> long
# Columns that identify each row
id_vars = ['GeoName', 'LineCode', 'IndustryClassification', 'Description', 'Unit']

# Columns to turn into rows
value_vars = [str(y) for y in range(2009, 2025)]

sagdp1_long = pd.melt(
    sagdp1_df,
    id_vars=id_vars,
    value_vars=value_vars,
    var_name='Year',
    value_name='DollarValue'
)

print('Melted successfully!')

print('DF shape:', sagdp1_long.shape)
print(sagdp1_long.info())
```

```python
# Convert LineCode from float64 to int
sagdp1_long['LineCode'] = sagdp1_long['LineCode'].astype(int)

# Convert Year from object to int
sagdp1_long['Year'] = sagdp1_long['Year'].astype(int)

print('LineCode and Year converted successfully!')

# Data check
print('Data check:')
print(sagdp1_long.info())
```

```python
# Strip whitespace from all text columns
text_cols = sagdp1_long.select_dtypes(include='object').columns

for col in text_cols:
    sagdp1_long[col] = sagdp1_long[col].str.strip()

print('Whitespace stripped successfully!')

# Reset index
sagdp1_long = sagdp1_long.reset_index(drop=True)

print('Index reset successfully!')
```

```python
# Check for duplicates
print('Number of duplicate rows:', sagdp1_long.duplicated().sum())
```

```python
# Save df to csv
sagdp1_long.to_csv('../../../workspace/nf-insights-project/data_clean/sagdp1_all_areas_clean.csv', index=False)

print('CSV saved successfully!')
```
