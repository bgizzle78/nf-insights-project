BEA SAGDP1 All Areas data cleaning and exploration.

```python
# Import libraries
import pandas as pd

print('Libraries imported successfully!')
```

```python
# Read in data
sagdp1_df = pd.read_csv('../../../workspace/nf-insights-project/data_raw/BEA GDP Data/SAGDP1__ALL_AREAS_1997_2024.csv', low_memory=False)

print('CSV read successfully!')

sagdp1_df.head()
```

```python
# Data check
print('Data check:')
print('DF shape:', sagdp1_df.shape)
print(sagdp1_df.info())
sagdp1_df.columns
```

```python
# Drop unnecessary columns
years = [str(y) for y in range(2008, 2025)]

sagdp1_df = sagdp1_df[['GeoName', 'LineCode', 'IndustryClassification', 'Description', 'Unit'] + years] 

print('Dropped unnecessary columns successfully!')

# Data check
print('Data check:')
print('Data shape:' ,sagdp1_df.shape)
sagdp1_df.info()
```

```python
# Drop completely empty rows
sagdp1_df = sagdp1_df.dropna(how='all', axis=0)

print('Completely empty rows dropped successfully!')

print('Data shape:', sagdp1_df.shape)
```

```python
# Keep only WV and US rows
sagdp1_df = sagdp1_df[sagdp1_df['GeoName'].isin(['West Virginia', 'United States'])]

print('Only WV and US rows kept successfully!')

print('DF shape:', sagdp1_df.shape)

sagdp1_df['GeoName'].unique()
```

```python
# Melt df from wide -> long
# Columns that identify each row
id_vars = ['GeoName', 'LineCode', 'Description', 'Unit']

# Columns to turn into rows
value_vars = [str(y) for y in range(2008, 2025)]

sagdp1_long = pd.melt(
    sagdp1_df,
    id_vars=id_vars,
    value_vars=value_vars,
    var_name='Year',
    value_name='GDPValue'
)

print('Data melted successfully!')

print('Data shape:', sagdp1_long.shape)
print(sagdp1_long.info())

sagdp1_long.head(20)
```

```python
# Convert LineCode from float64 to int64
sagdp1_long['LineCode'] = sagdp1_long['LineCode'].astype('Int64')

# Convert Year from object to int64
sagdp1_long['Year'] = sagdp1_long['Year'].astype('Int64')

print('Data types converted successfully!')

# Data check
print('Data check:')
print(sagdp1_long.info())
```

```python
# Select the columns we need to map industries
cols_to_check = ['LineCode', 'Description']

# Drop duplicates so we only see each industry once
bea_industries = sagdp1_long[cols_to_check].drop_duplicates()

# Sort by LineCode to see top-level sectors first
bea_industries = bea_industries.sort_values('LineCode')

# Show the rows
print(bea_industries.head(92).to_markdown(index=False))
```

```python
# Strip whitespace from all text columns
text_cols = sagdp1_long.select_dtypes(include='object').columns

for col in text_cols:
    sagdp1_long[col] = sagdp1_long[col].str.strip()

print('Whitespace stripped successfully!')
```

```python
# Check for duplicates
print('Number of duplicate rows:', sagdp1_long.duplicated().sum())

# Reset index
sagdp1_clean = sagdp1_long.reset_index(drop=True)

print('Index reset successfully!')
```

```python
# Save df to csv
sagdp1_clean.to_csv('../../../workspace/nf-insights-project/data_clean/sagdp1_all_areas_clean.csv', index=False)

print('CSV saved successfully!')
```
