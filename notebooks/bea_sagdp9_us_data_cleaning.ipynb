BEA SAGDP9 US data cleaning and exploration.

```python
# Import libraries
import pandas as pd

print('Libraries imported successfully!')
```

```python
# Read in data
sagdp9us_df = pd.read_csv('../../../workspace/nf-insights-project/data_raw/BEA GDP Data/SAGDP9_US_1997_2024.csv', low_memory=False)

print('CSV read successfully!')

sagdp9us_df.head()
```

```python
# Data check
print('Data check:')
print('Data shape:', sagdp9us_df.shape)
print(sagdp9us_df.info())
sagdp9us_df.columns
```

```python
# Drop unnecessary columns
years = [str(y) for y in range(2009, 2025)]

sagdp9us_df = sagdp9us_df[['GeoName', 'LineCode', 'IndustryClassification', 'Description', 'Unit'] + years]

print('Unnecessary columns dropped successfully!')

# Data check
print('Data shape:', sagdp9us_df.shape)
sagdp9us_df.info()
```

```python
# Drop completely empty rows
sagdp9us_df = sagdp9us_df.dropna(how='all', axis=0)

print('Completely empty rows dropped successfully!')

print('Data shape:', sagdp9us_df.shape)
```

```python
# Melt df from wide -> long
# Columns that identify each row
id_vars = ['GeoName', 'LineCode', 'IndustryClassification', 'Description', 'Unit']

# Columns to turn into rows
value_vars = [str(y) for y in range(2009, 2025)]

sagdp9us_long = pd.melt(
    sagdp9us_df,
    id_vars=id_vars,
    value_vars=value_vars,
    var_name='Year',
    value_name='GDPReal'
)

print('Data melted successfully!')

print('Data shape:', sagdp9us_long.shape)
print(sagdp9us_long.info())
```

```python
# Convert LineCode from float64 to int
sagdp9us_long['LineCode'] = sagdp9us_long['LineCode'].astype(int)

# Convert Year from object to int
sagdp9us_long['Year'] = sagdp9us_long['Year'].astype(int)

print('Data types converted successfully!')

# Data check
print('Data check:')
print(sagdp9us_long.info())
sagdp9us_long.head()
```

```python
# Select the columns we need to map industries
cols_to_check = ['LineCode', 'IndustryClassification', 'Description']

# Drop duplicates so we only see each industry once
bea_industries = sagdp9us_long[cols_to_check].drop_duplicates()

# Sort by LineCode to see top-level sectors first
bea_industries = bea_industries.sort_values('LineCode')

# Show the rows
print(bea_industries.head(92).to_markdown(index=False))
```

```python
# Strip whitespace from all text columns
text_cols = sagdp9us_long.select_dtypes(include='object').columns

for col in text_cols:
    sagdp9us_long[col] = sagdp9us_long[col].str.strip()

print('Whitespace stripped successfully!')
```

```python
# Pull out All industry total and Private industries
totals = sagdp9us_long[
    sagdp9us_long['Description'].isin(['All industry total', 'Private industries'])
][['Year', 'Description', 'GDPReal']]

# Aggregate totals by adding subsectors
subsectors = sagdp9us_long.dropna(subset=['IndustryName'])

grouped = subsectors.groupby(
    ['Year', 'IndustryName'],
    as_index=False
)['GDPReal'].sum()

# Rename totals column to match structure of grouped
totals = totals.rename(columns={'Description': 'IndustryName'})

# Concatenate totals and grouped
sagdp9us_grouped = pd.concat([totals, grouped], ignore_index=True)

# Data check
print('Data check:')
print('Data shape:', sagdp9us_grouped.shape)
sagdp9us_grouped.head()
```

```python
# Round gdp values
sagdp9us_grouped['GDPReal'] = sagdp9us_grouped['GDPReal'].round(1)

# Sort by Year and IndustryName
sagdp9us_grouped = sagdp9us_grouped.sort_values(
    ['Year', 'IndustryName']
)

# Data check
print('Data check:')
print('Data shape:', sagdp9us_grouped.shape)
sagdp9us_grouped.head()
```

```python
# Confirm summed subsector alignment with industry total
# Exclude the official total
components = sagdp9us_grouped[
    sagdp9us_grouped['IndustryName'].isin(
        ['All industry total', 'Private industries']
    )
]

# Sum components by year
component_sums = components.groupby('Year', as_index=False)['GDPReal'].sum()
component_sums = component_sums.rename(columns={'GDPReal': 'SumOfComponents'})
```

```python
# Get the official total
official_total = sagdp9us_grouped[
    sagdp9us_grouped['IndustryName'] == 'All industry total'
][['Year', 'GDPReal']]

official_total = official_total.rename(columns={'GDPReal': 'OfficialTotal'})
```

```python
# Merge and compare totals
comparison = pd.merge(component_sums, official_total, on='Year')

comparison['Difference'] = (
    comparison['OfficialTotal'] - comparison['SumOfComponents']
)

comparison['PercentDifference'] = (
    comparison['Difference'] / comparison['OfficialTotal'] * 100
)

comparison = comparison.round(4)

print(comparison.head())
```

```python
# Check for duplicates
print('Number of duplicate rows:', sagdp9us_grouped.duplicated().sum())

# Reset index
sagdp9us_clean = sagdp9us_grouped.reset_index(drop=True)

print('Index reset successfully!')
```
